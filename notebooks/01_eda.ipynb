# EDA: MMLU Dataset + Raw Model Responses

Explore the dataset structure, subject/domain balance, and raw model accuracy before any calibration analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mmlu = pd.read_parquet('../data/raw/mmlu.parquet')
print(mmlu.shape)
mmlu.head()
```

```python
# Questions per subject
mmlu.groupby(['domain', 'subject']).size().reset_index(name='n').sort_values('n')
```

```python
# Accuracy per subject (both models)
for model in ['gpt4o', 'llama']:
    try:
        df = pd.read_parquet(f'../data/processed/questions_{model}.parquet')
        acc = df.groupby('subject')['correct'].mean().sort_values()
        print(f'\n{model} accuracy range: {acc.min():.2f} - {acc.max():.2f}')
    except FileNotFoundError:
        print(f'{model}: not yet collected')
```
